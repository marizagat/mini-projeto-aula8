```python
#bibliotecas
import pandas as pd
```

```python
#bases de dados

'''A base de dados IMDb_movies traz informações acerca dos filmes.'''
movies = pd.read_csv('IMDb_movies.csv')

'''A base de dados IMDb_names traz informações acerca de atores e atrizes.'''
names = pd.read_csv('IMDb_names.csv')

'''A base de dados IMDb_ratings especifica melhor as informações sobre notas dos filmes.'''
ratings = pd.read_csv('IMDb_ratings.csv')
```
